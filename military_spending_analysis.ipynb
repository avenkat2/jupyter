```python
# DATS 6103 - Individual Project 1 - Poornima Joshi

import pandas as pd
import matplotlib.pyplot as plt


raw_data = pd.read_excel("SIPRI-Milex-data-1949-2017.xlsx", sheet_name="Constant (2016) USD")
raw_data.columns = raw_data.iloc[4]
raw_data.head()
```

```python
raw_data = raw_data.drop([0,1,2,3,4])
raw_data = raw_data.drop(columns = ['Notes'])
raw_data =  raw_data.set_index('Country')
```

```python
country_list = ['USA','China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Israel']
df = raw_data.loc[country_list]
df
```

```python
col_list = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
df = df[col_list]
#df = df.reset_index()
df
```

```python

for i in df.index:
    plt.plot(df.loc[i])

plt.xlabel('Year')
plt.ylabel('Military spending in USD')
plt.title('Military spending of several countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df1 = df.copy()
# Compute Percentages
df2 = df1.apply(lambda x : x/x.sum() * 100)
# Transpose to get YoY data
df3 = df2.T
#Visualize area chart to check the countries with the biggest budget YoY
print(df3)
df3.plot.area()
```

```python
colors = ['Red', 'Plum', 'Yellow', 'Green', 'Blue', 'Orange', 'Purple', 'Gold',
          'Magenta', 'Grey']
ax = df3.plot( kind='bar', width= .6, figsize = (25,15), color = colors, fontsize = 16, 
                 title = 'Military spending of Several countries')
plt.legend(loc = 'best')
ax.set_xlabel('Year', fontsize = 22)
ax.set_ylabel('Military Spending [Millions of US Dollars])', fontsize = 20)
```

```python
raw_data_gdp = pd.read_excel("SIPRI-Milex-data-1949-2017.xlsx", sheet_name="Share of GDP")
raw_data_gdp.columns = raw_data_gdp.iloc[4]
raw_data_gdp.head()
raw_data_gdp = raw_data_gdp.drop([0,1,2,3,4])
print(raw_data_gdp.head())
raw_data_gdp = raw_data_gdp.drop(columns = ['Notes'])
raw_data_gdp =  raw_data_gdp.set_index('Country')
country_list = ['USA','China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Israel']
gdp = raw_data_gdp.loc[country_list]
col_list = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
print(gdp[col_list])
gdp = gdp[col_list] * 100
#df = df.reset_index()
gdp.T.plot.area()
```

```python
ax1 = gdp.T.plot( kind='bar', width= .6, figsize = (25,15), color = colors, fontsize = 16, 
                 title = 'Military spending of Several countries to percentage of GDP')
plt.legend(loc = 'best')
ax1.set_xlabel('Year', fontsize = 22)
ax1.set_ylabel('Percentage of GDP', fontsize = 20)
```
